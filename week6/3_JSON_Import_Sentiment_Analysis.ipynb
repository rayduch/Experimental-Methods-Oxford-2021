```python
import pandas as pd

df = pd.read_json(r'C:\Users\Macie\Jupyter\Ray Duch\Sentiment Analysis\vaccine_tweets_testrun.json', lines = True)
df
```

```python
from afinn import Afinn
import nest_asyncio
from nltk.corpus import stopwords
stop = stopwords.words('english')
nest_asyncio.apply()

afinn = Afinn()
nyt_df = df

nyt_df['tweet_without_stopwords'] = nyt_df['tweet'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
nyt_df
```

```python
nyt_df.tweet_without_stopwords.str.split(expand=True).stack().value_counts()
```

```python
from afinn import Afinn
nyt_df['afinn_score'] = nyt_df['tweet'].apply(afinn.score)
```

```python
nyt_df['afinn_score'].describe()
```

```python
columns_to_display = ['tweet', 'afinn_score']
pd.set_option('max_colwidth', 1000)

nyt_df.sort_values(by='afinn_score')[columns_to_display].head(10)
```

```python
def word_count(text_string):
    '''Calculate the number of words in a string'''
    return len(text_string.split())
```

```python
nyt_df['word_count'] = nyt_df['tweet'].apply(word_count)
nyt_df['word_count'].describe()
```

```python
nyt_df['afinn_adjusted'] = nyt_df['afinn_score'] / nyt_df['word_count'] * 100
nyt_df['afinn_adjusted'].describe()
```

```python
analysis = nyt_df.groupby('username')['afinn_adjusted'].describe()
```

```python
vaccines_test_df = analysis.sort_values('count', ascending = False)
```

```python
vaccines_test_df.to_excel('vaccines_test_df.xlsx')
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

analyzer.polarity_scores('Horrible bad day.')
```

```python
sentiment = nyt_df['tweet'].apply(analyzer.polarity_scores)
sentiment
```

```python
sentiment_df = pd.DataFrame(sentiment.tolist())

sentiment_df.head()
```

```python
nyt_df_sentiment = pd.concat([nyt_df['tweet'],nyt_df['username'], sentiment_df], axis = 1)
nyt_df_sentiment
```
