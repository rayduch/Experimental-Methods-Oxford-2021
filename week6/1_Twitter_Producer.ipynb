### 3rd step: Configure the StreamListener to our topic and let it run!

```python
class StdOutListener(StreamListener):
    def on_data(self, data):
        producer.send("livetrump", data.encode('utf-8'))
        
        # Print out the twitter data (use with caution as it will cause the notebook to become unresponsive)
        print (data)

        return True
    
    # Called when a non-200 status code is returned and print the status message
    def on_error(self, status):
        print(status)
```
